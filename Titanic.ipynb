```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(),cbar=False ,cmap='viridis')
plt.title('NA Heatmap')
plt.show()
```

```python
df['Age_Missing'] = df['Age'].isnull().astype(int)
```

```python
df
```

```python
df['Embarked_Missing'] = df['Embarked'].isnull().astype(int)
```

```python
df
```

```python
df = df.drop('Cabin',axis =1)
```

```python
df
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(),cbar=False ,cmap='viridis')
plt.title('NA Heatmap')
plt.show()
```

```python
df['Embarked']=df['Embarked'].fillna('S')
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(),cbar=False ,cmap='viridis')
plt.title('NA Heatmap')
plt.tight_layout()
plt.savefig("na_heatmap.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Feature Distribution (Boxplot)')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```

```python
Q1 = df['Fare'].quantile(0.25)
Q2 = df['Fare'].quantile(0.75)
IQR = Q1- Q2
```

```python
lower_bound = Q1-1.5*IQR
upper_bound = Q2+1.5*IQR
```

```python
df['Fare']=np.where(df['Fare']>lower_bound,lower_bound,df['Fare'])
df['Fare']=np.where(df['Fare']<upper_bound,upper_bound,df['Fare'])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Feature Distribution (Boxplot)')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```

```python
df.to_csv("clean_v1.csv", index=False)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Fare Distribution by Passenger Class")
plt.tight_layout()
plt.savefig("fare_vs_pclass.png")
plt.show()
plt.close()
```
